```python
import math
import nltk
import re
## nltk settings
nltk.download('punkt')
```

```python
def ReadData():  # Read From File
    url = "A:\\4 second term\\NLP\Assignments\\Assignment 1\\trigramemodel\\Dataset\\data.txt"
    with open(url, "r", encoding="utf8") as f:
        data = f.read()
    return data
data = ReadData()
```

```python
def cleaningData(data):
    data = re.sub('[^A-Za-z0-9\s]+', '', data)
    sentences = data.split('\n')

    sentences = [s.strip() for s in sentences]
    sentences = [s for s in sentences if len(s) > 0]
    return sentences
```

```python
def preprocessing(data):
    tokens = []
    finalTokens = []

    # cleaning Data
    sentences = cleaningData(data)

    for sentence in sentences:
        sentence = sentence.lower()
        token = nltk.word_tokenize(sentence)
        tokens.append(token)  # [[],[],[]]

    for words in tokens:
        for word in words:
            finalTokens.append(word)
    # print(finalTokens)
    return finalTokens  # [,,,,,]
tokens = preprocessing(data)
```

```python
def ngrame(data, n):
    ngrame = []
    for i in range(len(data)):
        D = data[i: i + n]
        size = len(data[i: i + n])
        if size == n:
            ngrame.append(D)
    return ngrame
```

```python
unigram = ngrame(tokens, 1)
bigram = ngrame(tokens, 2)
trigram = ngrame(tokens, 3)
```

```python
def count_the_words(data):
    frequency = {}
    for t in data:
        if tuple(t) in frequency:
            frequency[tuple(t)] += 1
        else:
            frequency[tuple(t)] = 1
    return frequency
```

```python
unigram = count_the_words(unigram)
bigram = count_the_words(bigram)
trigram = count_the_words(trigram)
```

```python
def probability(unigram, bigram, trigram):
    model = []
    for trigram_value, trigram_count in trigram.items():
        prob = 1

        prob *= unigram[(trigram_value[0],)]
        prob *= bigram[(trigram_value[0], trigram_value[1])] / unigram[(trigram_value[0],)]
        prob *= trigram_count / bigram[(trigram_value[0], trigram_value[1])]

        model.append([trigram_value, prob / len(trigram)])
    return model
```

```python
model = probability(unigram,bigram,trigram)
```

```python
def suggestion(input):
    suggest = []
    input = input.lower()
    for p, v in model:
        if p == '_':
            continue
        for i in range(len(p)):
            if not (p[i].startswith(input)):
                continue
            arr = []
            if i == 0:
                arr.append(p[0])
                arr.append(p[1])
                arr.append(p[2])
            elif i == 1:
                arr.append(p[1])
                arr.append(p[2])
            if len(arr):
                suggest.append((arr, v))
    return list(filter(suggest))
```

```python
def filter(suggest):
    filterd = []
    temp = []
    final_suggest = []
    for i in suggest:
        if temp.count(i[0]) == 0:
            filterd.append(i)
            temp.append(i[0])
        else:
            tup = [item for item in filterd if i[0] in item]
            if i[1] > tup[0][1]:
                index = filterd.index(tup[0])
                filterd[index] = i
    filterd.sort(key=lambda a: a[1], reverse=True)
    for i in filterd:
        arr = []
        for j in i[0]:
            arr.append(j)
        final_suggest.append(arr)
    return final_suggest
```

```python
Allsuggestion=suggestion("do")
```
